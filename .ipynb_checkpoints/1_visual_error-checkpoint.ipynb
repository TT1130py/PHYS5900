## Example 1: Contour/density plots

Used to visualize two-dimensional data, e.g. two-dimensional functions $f(x,y)$

For example, the electric potential of a dipole reads (dimensionless units)

$V_{dip} (x,y) = \frac{1}{\sqrt{(x-x_1)^2+(y-y_1)^2}} - \frac{1}{\sqrt{(x-x_2)^2+(y-y_2)^2}}$

Here $(x_1,y_1)$ and $(x_2,y_2)$ are the coordinates of the poles

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------------------
# Step 1: Define the grid of points where we want to compute the potential
# -------------------------------------------------------------
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

# -------------------------------------------------------------
# Step 2: Define the positions of the charges that form our dipole
# -------------------------------------------------------------
# Let's put +q at (-2, 0) and -q at (+2, 0).
x1, y1 = -2, 0   # position of +q
x2, y2 = 2, 0   # position of -q

# -------------------------------------------------------------
# Step 3: Build a 2D meshgrid
# -------------------------------------------------------------
# Meshgrid takes the 1D x and y arrays and produces 2D arrays X and Y
# such that every combination of (x, y) appears once. This is needed
# for evaluating functions over a whole 2D region.
X, Y = np.meshgrid(x, y)

print(X.shape, Y.shape)

print(Y)
# -------------------------------------------------------------
# Step 4: Compute the electric potential
# -------------------------------------------------------------

#Use capital X and Y 
Vdip1 = 1.0 / np.sqrt(((X-x1)**2)+((Y-y1)**2)) 
#print(Vdip1)
Vdip2 = 1.0 / np.sqrt(((X-x2)**2)+((Y-y2)**2))
#print(Vdip2)
VdipF = Vdip1 - Vdip2
print(VdipF)
```

```python
# -------------------------------------------------------------
# Step 5: Plot equipotential lines (contour plot)
# -------------------------------------------------------------
# A contour plot shows lines of constant potential, like a topographic map.
# levels = np.linspace(-0.5, 0.5, 11) means: draw contour lines
# at equally spaced values between -0.5 and +0.5.
plt.figure()
CS = plt.contour(X, Y, VdipF, levels=np.linspace(-0.5, 0.5, 11))
#plt.clabel(CS, inline=True, fontsize=8)  # label the contours
plt.title("Electric potential of a dipole (contour lines)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# -------------------------------------------------------------
# Step 6: Plot filled contours (color map)
# -------------------------------------------------------------
# Contourf (filled contours) shades the regions between levels.
# This makes it easier to see high and low potential regions at a glance.
fig2, ax2 = plt.subplots(constrained_layout=True)
CS2 = ax2.contourf(X, Y, VdipF, levels=np.linspace(-0.5, 0.5, 11), cmap="RdBu")
ax2.set_title("Electric potential of a dipole (filled contours)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_aspect("equal")
#fig2.colorbar(CS2, ax=ax2, label="Potential (arbitrary units)")
plt.show()
```

```python
#For each line on the contour plot- every x,y along the line has the same electric potential
```

## Example 2: Errors in numerical derivative

Consider a function 
$$f(x) = x(x-1)~.$$
Its derivative is 
$$f'(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h} = 2x - 1$$

Let us calculate the derivative numerically by using small but finite values of $h$ ranging from $1$ down to $10^{-16}$ at $x = 1$
and compare it to the correct result, $f'(1) = 1$.

```python
def f(x):
    return x*(x-1.)

def df_exact(x):
    return 2.*x - 1.

def df_numeric(x,h):
    return (f(x+h) - f(x))/h #this is correct

print("{:<10} {:<20} {:<20}".format('h',"f'(1)","Relative error"))
x0 = 1.

arr_h = []
arr_df = []
arr_err = []

#0 to -20 with step size -1
for i in range(0,-20,-1):
    h = 10**i
    df_val = df_numeric(x0, h)
    df_err = abs(df_numeric(x0,h) - df_exact(x0)) / df_exact(x0)
    print("{:<10} {:<20} {:<20}".format(h,df_val,df_err))
    arr_h.append(h)
    arr_df.append(df_val)
    arr_err.append(df_err)
```

Let us plot the dependence of the relative accuracy vs $h$

```python
import matplotlib.pyplot as plt

params = {'legend.fontsize': 'large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'xtick.direction':'in',
         'ytick.direction':'in',
         }
plt.rcParams.update(params)

plt.title("Accuracy of the numerical derivative")
plt.xlabel("${h}$", fontsize=18)
plt.ylabel("relative error", fontsize=18)
plt.xscale('log')
plt.yscale('log')
plt.scatter(arr_h, arr_err, color="red")

plt.show()
```
